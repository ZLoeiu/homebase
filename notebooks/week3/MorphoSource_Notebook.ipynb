## MorphoSource

MorphoSource is a digital repository hosted by Duke University Research Computing and aggregates a collection of 3D images and media pertaining to physical objects and/or biological specimens. Researchers, museum curators, universties, organzations, and the general public are able to download and interact with these data files through analysis and/or personal exploration. 

Personally, I think that the available images are really cool to look at, especially since most of them are scanned and uploaded on 3D platforms so that users can really look at the specimens/objects in various perspectives and orientations. Go check it out! Currently, MorphoSouce is planning on improving the diversity and complexity of the 3D images preserved and the security of the datasets stored on their platform. More information can be found on their [About Me](https://www.morphosource.org/About/home) page!

## Become a MorphoSource User

In order to download data from MorphoSource directly, you must create an account. It's free! And you can sign up [here](https://www.morphosource.org/LoginReg/form).

If you need a particular species we will move the slice data here ASAP. Email Adam Summers. This work is supported by the National Science Foundation and the Seaver Institute."

You can access this dataset [here](https://www.morphosource.org/Detail/ProjectDetail/Show/project_id/220).

For the purposes of this example, I will be doing things on a Jupiter notebook, and importing modules like numpy and Berkeley's datascience package (but you can totally do this with pandas or something else too!)

The first thing you should do is import all these packages/modules in your cell on Jupiter

```python
from datascience import *
import numpy as np
```

Then, make sure that you have uploaded your csv file that you downloaded. When you have done that, you can directly read the csv file and you will be able to see your data! (aka type in the following in the next cell)

Note: Your file name inside the method "read_table" should be the name of your csv file (it may not necessarily be "Fish_scans")

```python
file = Table().read_table("Fish_scans.csv")
file
```

```python
filtered_table = file.select("Scanner", "family", "Species", "Genus", "scan date")
filtered_table
```

If I wanted to look specifically at only the fishes scanned from the Bruker Skyscan 1173 scanner, I can do that too!

```python
BrukerSkyscan = filtered_table.where("Scanner", "Bruker Skyscan 1173")
BrukerSkyscan
```

If I want to look at all of the data, I can just put ".show()" at the end of my variable name, and I will be able to see all of my data, and not just the first 10. (Do note that this can take a while if the dataset is extremely large)

```python
# BrukerSkyscan.show()
```

Let's group the fish by their Genus and create a simple bar chart visualization. 

```python
GenusGrouped = BrukerSkyscan.group("Genus")
GenusGrouped
# BrukerSkyscan.where("Genus","#VALUE!") // you can see the fish that don't have Genus here (just uncomment)
```

Hmm... it seems that nearly 368 of these fish have an unknown or unidentified Genus. Perhaps that data has not be provided yet, but for now, we have two options, either ignore this data or keep it. Since 368 seems like a rather large amount of data to ignore, we should keep it. But because the data set is so big, I'm only going to take every 10th row up to the 100th row, and create a bar graph from the selected data. 

Graphing it below, we now have a bar chart visualization for all the fish scanned by Bruker Skyscan 1173 grouped by Genus. 

```python
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

filteredGenus = GenusGrouped.take(np.arange(10,100))
filteredGenus

filteredGenus.barh("Genus")
```

From this, we can see that fish from our systematically selected rows/data are mainly from the Agonopsis and Anodontostoma Genus.
